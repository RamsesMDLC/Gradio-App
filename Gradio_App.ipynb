```python
import gradio as gr
```

```python
#Build an UI based on the function "greet" which has 2 inputs, a "name" and the property "intensity" (i.e., repetition or duplication)
def greet(name, intensity):
    return "Hello, " + name + "!" * int(intensity)

#Properties of the UI
  #Function, input ()"text" for "name" and "slider" for "intensity"
demo = gr.Interface(
    fn=greet,
    inputs=["text", "slider"],
    outputs=["text"],
    api_name="predict"
)

demo.launch()
```

## **Test 3**

```python
def greet(name, is_morning, temperature):
  salutation = "Good morning" if is_morning else "Good evening"
  greeting = f"{salutation} {name}. It is {temperature} degree today"
  celsius = (temperature - 32)*5/9
  return greeting, round(celsius,2)

demo = gr.Interface(
    fn=greet,
    inputs=["text", "checkbox", gr.Slider(0,100)],
    outputs=["text", "number"],
    api_name="predict"
)

demo.launch()
```

## **Test 4**

```python
import numpy as np
import gradio as gr

def sepia(input_img):
  sepia_filter = np.array([
      [0.393,0.769,0.189],
      [0.349,0.686,0.168],
      [0.272,0.534,0.131]
  ])
  sepia_img = input_img.dot(sepia_filter.T)
  sepia_img /= sepia_img.max()
  print(input_img.shape, sepia_img.shape)
  return sepia_img

demo = gr.Interface(sepia, gr.Image(),"image")

demo.launch()
```

```python
import gradio as gr

from transformers import pipeline

pipe = pipeline("translation", model = "Helsinki-NLP/opus-mt-it-en")
```

```python
def translate(text):
  return pipe(text)[0]["translation_text"]

with gr.Blocks() as demo:
  with gr.Row():
    with gr.Column():
      italian = gr.Textbox(label = "Italian Text")
      translate_btn = gr.Button(value = "Translate")
    with gr.Column():
      english = gr.Textbox(label = "English Text")
  translate_btn.click(translate, inputs = italian, outputs = english)
  examples = gr.Examples(examples =["donna", "woman"], inputs = [italian])


demo.launch(share = True)
```
